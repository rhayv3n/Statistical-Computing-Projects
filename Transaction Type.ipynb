```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data_path = r"C:\Users\Rhayven\Downloads\archive\ffsales.csv"
df = pd.read_csv(data_path)
```

```python
df['transaction_type'].fillna('Others', inplace=True)
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='transaction_type')
plt.title('Distribution of Transaction Types')
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='transaction_type', y='transaction_amount')
plt.title('Impact of Transaction Types on Total Revenue')
plt.xlabel('Transaction Type')
plt.ylabel('Total Revenue')
plt.show()
```

```python
total_revenue = df['transaction_amount'].sum()

transaction_type_revenue = df.groupby('transaction_type')['transaction_amount'].sum()
contribution_percent = (transaction_type_revenue / total_revenue) * 100

print("Total Revenue:", total_revenue)
print("\nRevenues Distributed:")
print(transaction_type_revenue)
print("\nContribution of each transaction type (Percent):")
print(contribution_percent)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data_path = r"C:\Users\Rhayven\Downloads\archive\ffsales.csv"
df = pd.read_csv(data_path)

# Replace missing values in 'transaction_type' with 'Others'
df['transaction_type'].fillna('Others', inplace=True)

# Create a bar plot to visualize the distribution of transaction types based on total revenue
plt.figure(figsize=(10, 6))
sns.barplot(data=df, x='transaction_type', y='transaction_amount', estimator=sum, errorbar=None)
plt.title('Distribution of Transaction Types by Total Revenue')
plt.xlabel('Transaction Type')
plt.ylabel('Total Revenue')
plt.show()

# Create a box plot to analyze the impact of transaction types on total revenue
plt.figure(figsize=(10, 6))
sns.boxplot(data=df, x='transaction_type', y='transaction_amount')
plt.title('Impact of Transaction Types on Total Revenue')
plt.xlabel('Transaction Type')
plt.ylabel('Total Revenue')
plt.show()

```
